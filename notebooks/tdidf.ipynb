```python
colecao_medicina = fetch_20newsgroups(subset='train', categories=['sci.med'], remove=('headers', 'footers', 'quotes'))
colecao_query_medicina = fetch_20newsgroups(subset='test', categories=['sci.med'], remove=('headers', 'footers', 'quotes'))
```

```python
lema_subset = [preprocess_lemma(doc) for doc in colecao_medicina.data]
```

```python
usando_data = lema_subset
usando_query = query_lema_subset
```

```python
vocabulary, term_to_idx = build_vocabulary(usando_data)
print(f"Vocabulário: {len(vocabulary)} termos")
print(vocabulary[:5])
print(term_to_idx)
```

```python
# --- Passo 1: Calcular DF ---
df_vector = calcular_frequencia_documento(usando_data, vocabulary, term_to_idx)
print(f"\nDF para os 5 primeiros termos: {df_vector[:5]}")

# --- Passo 2: Calcular IDF ---
idf_vector = calcular_idf(df_vector, total_documentos=len(usando_data))
print(f"\nIDF para os 5 primeiros termos: {idf_vector[:5]}")

# --- Passo 3: Construir as Matrizes Finais ---
tf_matrix, tfidf_matrix = calcular_tf_idf_matriz(usando_data, vocabulary, term_to_idx, idf_vector)
print(f"\nFormato da matriz TF-IDF: {len(tfidf_matrix)} documentos x {len(vocabulary)} colunas.")
```

```python
query = usando_query
query_tokens = query[0]  # Processa a primeira query como exemplo
query_vector = calcular_tf_idf_query(query_tokens, vocabulary, term_to_idx, idf_vector)
for query_term, value in zip(vocabulary, query_vector):
    if value > 0:
        print(f"Termo: '{query_term}', TF-IDF: {value:.4f}")
```
